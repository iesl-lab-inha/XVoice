# 22.05.05 Simple Background Noise Reduction by Min
## Used recorded voice audio saying "Two plus seven is less than ten" by myself.
## In the original audio, you can hear a lot of hiss or buzz sounds which is the background noise when the voice was recorded.
## After using noise reduction provided by https://pypi.org/project/noisereduce/ you can get clear voice audio without the background noise.

```python
import librosa
import librosa.display
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Audio, IFrame, display
import noisereduce as nr
```

```python
sr = 16000
e_file1 = 'myvoice.wav'
y1,sr = librosa.load(e_file1, mono=True, sr=sr)
```

```python
display(Audio(y1,rate=sr))
```

```python
librosa.display.waveshow(y1,sr=sr, x_axis='time');
```

```python
# perform noise reduction
reduced_noise = nr.reduce_noise(y=y1, sr=sr)
```

```python
librosa.display.waveshow(reduced_noise,sr=sr, x_axis='time');
```

```python
display(Audio(reduced_noise,rate=sr))
```

```python
display("Original Audio", Audio(data=y1, rate=sr))
display("Noise Reduced Audio", Audio(data=reduced_noise, rate=sr))
```
